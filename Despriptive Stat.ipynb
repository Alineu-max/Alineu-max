```python
#libraries
import pandas as pd
import seaborn as sns
import  scipy.stats as st
```

```python
#load the dataset
df = pd.read_csv("Baseball.csv")
df.head()
```

```python
#info dataset
df.info()
```

```python
#Mean of run scored
df.RS.mean()
```

```python
# mean of runs scored by arizona
df.loc[df.Team == "ARI"]
```

```python
df.loc[df.Team == "ARI"].RS.mean()
```

```python
#mean of runs scored by arizona team since 2005
df.loc[(df.Team == "ARI") & (df.Year >= 2005)].RS.mean()
```

```python
#mean of runs scored by arizon chicago team since 2000- 2005
df.loc[(df.Team == "CHC") & ((df.Year < 2000)| (df.Year > 2005))].RA.mean()
```

```python
# mean and meadian of wins
print(df.W.mean())
print(df.W.median())
```

```python
#mean of Baltimore Team until 2000
df.loc[(df.Team == "BAL") & (df.Year <= 2000)].W.median()
```

```python
# mean. mode and median
print(df.OBP.mean())
print(df.OBP.mode())
print(df.OBP.median())
df.OBP.hist()
```

```python
# mode of OBP of the year 2010
df.loc[df.Year == 2010].OBP.mean()
```

```python
# most runk season in Boston
df.loc[df.Team == "BOS"].RankSeason.mode()

```

Standard Deviation and variance


```python
#OOBP
print(df.OOBP.mean())
print(df.OOBP.std())
print(f"mean: {df.OOBP.mean()}")
print(f"STD sample: {df.OOBP.std()}")
print(f"STD population: {df.OOBP.std()}")
print(f"STD variance: {df.OOBP.var()}")
```

```python
#BA
print(df.BA.mean())
print(df.BA.std())
print(f"mean: {df.BA.mean()}")
print(f"STD sample: {df.BA.std()}")
print(f"STD population: {df.BA.std()}")
print(f"STD variance: {df.BA.var()}")
```
